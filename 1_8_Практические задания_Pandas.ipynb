```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("my_stat_1.csv")
data
```

```python
med=data.n_users.loc[data.n_users>0].median()
n_users=data.n_users
data.n_users=n_users.replace(list(n_users[n_users<0]),med)
data.session_value=data.session_value.fillna(0)
```

```python
data
```

```python
mean_session_value_data=data.groupby('group',as_index=False 

).agg({'session_value': 'mean'}).rename(columns={"session_value":"mean_session_value"})
```
